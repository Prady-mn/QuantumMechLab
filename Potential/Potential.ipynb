```python
def plotEnergy(n = [1,2,3],w = 1e-3,A = 1):
    n = list(n)
    
    fig = plt.figure(figsize = (10,5))
    X = np.linspace(-A,A,100)
    plt.plot(X,[V(x,w) for x in X],color = "black")
    plt.xlim(-A,A)
    
    color_dict = dict(zip(n,random.sample(list(mcolors.BASE_COLORS.values())[:-2],k = len(n))))
    for i in n:
        plt.axhline(energy(i,w),color = color_dict[i],label = f"n = {i}")

    
    plt.grid(axis = 'y',which = 'both')

    xloc = plt.xticks()[0]
    plt.xticks(ticks = xloc, labels = [f"{Fraction(i)}A" for i in xloc])
    plt.legend(title = f"A = {A}, w = {w:.2e}",fontsize = 10)
    #plt.savefig("EnergyDiagram")
    plt.show()

widgets.interactive(plotEnergy,
                    n = widgets.SelectMultiple(
                        options=[1,2,3,4,5],
                        value=[1],
                        rows=5,
                        description='n = ',
                        disabled=False),
                    w = widgets.FloatSlider(
                        value = 1e-3,
                        min = 0.5e-3,
                        max = 2e-3,
                        step = 0.25e-3,
                        description = "w = ",
                        readout_format = '.2e'),
                    A = widgets.fixed(1))
```
